```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bar
barWidth = 0.25
fig,ax = plt.subplots(figsize =(20, 10))

# set height of bar
full_size_training = [77.80, 78.69, 79.93, 79.56, 79.75, 79.38]
transition_training = [52.68, 53.75, 53.99, 50.81, 56.67, 50.74]

# Set position of bar on X axis
br1 = np.arange(len(full_size_training))
br2 = [x + barWidth for x in br1]
# br3 = [x + barWidth for x in br2]

# Make the plot
full = ax.bar(br1, full_size_training, color ='r', width = barWidth, edgecolor ='grey', label ='Full Size Training Data')
for p in full:
   height = p.get_height()
   ax.annotate('{}'.format(height),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')
transition = ax.bar(br2, transition_training, color ='b', width = barWidth, edgecolor ='grey', label ='Transition Zone Training Data')
for p in transition:
   height = p.get_height()
   ax.annotate('{}'.format(height),
      xy=(p.get_x() + p.get_width() / 2, height),
      xytext=(0, 3), # 3 points vertical offset
      textcoords="offset points",
      ha='center', va='bottom')

# Adding Xticks
ax.set_xlabel('Overlapping Window Size', fontweight ='bold', fontsize = 15)
ax.set_ylabel('Accuracy %', fontweight ='bold', fontsize = 15)
ax.set_xticks(br1)
ax.set_xticklabels(['100ms', '150ms', '200ms', '250ms', '350ms', '500ms'])

ax.set_ylim(0, 100)
ax.set_title('Testing Accuracy for Variant Window (data from subject s010)',size=30)
ax.legend()
plt.show()
```
