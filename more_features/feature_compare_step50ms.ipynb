# test feature performance

## STD

```python
name = 'STD'

trainx = pd.read_csv('featured_data/step50ms/'+name+'_trainx.csv')
trainx = trainx.drop(columns=['Unnamed: 0'])
trainy = pd.read_csv('featured_data/step50ms/'+name+'_trainy.csv')
trainy = trainy.drop(columns=['Unnamed: 0'])
testx = pd.read_csv('featured_data/step50ms/'+name+'_testx.csv')
testx = testx.drop(columns=['Unnamed: 0'])
testy = pd.read_csv('featured_data/step50ms/'+name+'_testy.csv')
testy = testy.drop(columns=['Unnamed: 0'])

# train/test model
lightGBM_model = ltb.LGBMClassifier()
lightGBM_model.fit(trainx,trainy)
expected1 = testy.values.flatten()
predicted1 = lightGBM_model.predict(testx)
counter = 0
for i in range(0, len(expected1)):
    if expected1[i] == predicted1[i]:
        counter += 1

accuracy = counter/len(expected1)

acc_features.append(accuracy)
name_features.append(name)
```

# feature cross

## combine features for train test data

```python
selected_features = ['STD', 'RMS', 'IEMG', 'MAV', 'WL', 'SSI', 'AAC', 'DASDV']
```

```python
name = 'STD'
trainx = pd.read_csv('featured_data/step50ms/'+name+'_trainx.csv')
trainx = trainx.drop(columns=['Unnamed: 0'])
combine_trainx = trainx

for name in selected_features:
  if name == 'STD':
    continue
  
  trainx = pd.read_csv('featured_data/step50ms/'+name+'_trainx.csv')
  trainx = trainx.drop(columns=['Unnamed: 0'])
  combine_trainx = pd.concat([combine_trainx,trainx],axis=1)

name = 'STD'
testx = pd.read_csv('featured_data/step50ms/'+name+'_testx.csv')
testx = testx.drop(columns=['Unnamed: 0'])
combine_testx = testx

for name in selected_features:
  if name == 'STD':
    continue
  
  testx = pd.read_csv('featured_data/step50ms/'+name+'_testx.csv')
  testx = testx.drop(columns=['Unnamed: 0'])
  combine_testx = pd.concat([combine_testx,testx],axis=1)

combine_trainy = pd.read_csv('featured_data/step50ms/'+name+'_trainy.csv')
combine_trainy = combine_trainy.drop(columns=['Unnamed: 0'])

combine_testy = pd.read_csv('featured_data/step50ms/'+name+'_testy.csv')
combine_testy = combine_testy.drop(columns=['Unnamed: 0'])
```

```python
# train/test model
lightGBM_model = ltb.LGBMClassifier()
lightGBM_model.fit(combine_trainx,combine_trainy)
expected1 = combine_testy.values.flatten()
predicted1 = lightGBM_model.predict(combine_testx)
counter = 0
for i in range(0, len(expected1)):
    if expected1[i] == predicted1[i]:
        counter += 1

accuracy = counter/len(expected1)
```
